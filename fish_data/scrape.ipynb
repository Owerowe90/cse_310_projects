```python
import time
import urllib.request
import os
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# Rainbow Trout
url = "https://www.google.com/search?q=rainbow+trout+fish&source=lnms&tbm=isch&sa=X&ved=2ahUKEwjlzJ2To9j-AhUNJkQIHRSNDgkQ0pQJegQIBRAC&biw=1470&bih=764&dpr=2"

# Initialize a web driver
driver = webdriver.Chrome()

# Send a GET request to the URL
driver.get(url)

# Find the first image tag in the HTML content
image_tags = driver.find_elements(By.TAG_NAME, "img")
first_image_count = len(image_tags)

# Scroll down to load more images dynamically
scroll_down_count = 0
while True:
    # Scroll down
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1)
    scroll_down_count += 1
    print(scroll_down_count)

    # Find all the image tags in the HTML content
    image_tags = driver.find_elements(By.TAG_NAME, "img")

    # Download the new images
    folder_path = "/Users/owenrowe/Desktop/cse310/fish_data/rainbow/"
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    for i, tag in enumerate(image_tags[first_image_count:]):
        try:
            image_url = tag.get_attribute("src")
            if image_url.startswith("http"):
                filename = folder_path + "image_" + str(first_image_count + i + 1) + ".jpg"
                urllib.request.urlretrieve(image_url, filename)
        except:
            pass

    if scroll_down_count == 10:
        break

# Close the browser
driver.quit()
```
