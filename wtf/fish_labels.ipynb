```python
main_folder = '/content/gdrive/MyDrive/WTF/fish_data'
csv_file = '/content/gdrive/MyDrive/WTF/fish_data/fish_image_labels.csv'

# Open the CSV file in write mode
with open(csv_file, 'w', newline='') as file:
    writer = csv.writer(file)

    # Write header row
    writer.writerow(['image', 'class'])

    # # Iterate through the image names using a for loop
    # for image_name in image_names:
    #     # Extract the file name without extension
    #     file_name = image_name.split('.')[0]

    #     # Write the image name and file name to the CSV file
    #     writer.writerow([image_name, file_name])


    # Iterate through subfolders and images
    for folder_name in os.listdir(main_folder):
        folder_path = os.path.join(main_folder, folder_name)
        if not os.path.isdir(folder_path):
            continue  # Skip if not a subfolder
        
        file_num = 1
        print(f"Processing images in {folder_name}...")

        for file_name in os.listdir(folder_path):
            if file_name.startswith('.'):
                continue  # Skip hidden files
            
            # Split file_name into name and extension
            file_name_parts = os.path.splitext(file_name)
            file_extension = file_name_parts[1]
            new_file_name = f"{folder_name}_{file_num}{file_extension}"

            file_path = os.path.join(folder_path, file_name)
            new_file_path = os.path.join(folder_path, new_file_name)
            
            # Rename the file
            os.rename(file_path, new_file_path)
            file_num += 1

            # Write the image name and file name to the CSV file
            writer.writerow([new_file_name, folder_name])
```
